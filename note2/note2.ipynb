```python
# define the system

def d_system_2(t, y, a, b, c, d, e, f):
    x, y_ = y
    dx_dt = x * (a - b * y_ - e * x)
    dy_dt = y_ * (-c + d * x - f * y_)
    return [dx_dt, dy_dt]
```

```python
# set parameters

a, b, c, d, e, f = 1, 2, 3, 4, 0.5, 0.5
params = (a, b, c, d, e, f)
```

```python
# define time span and initial conditions

t_span = (0, 50)
initial_conditions = [
    [0.5, 0.5],
    [1, 1],
    [1.5, 1.5],
    [2, 2]
]
```

```python
# plot the trajectories

plt.figure(figsize=(8, 6))
for init_cond in initial_conditions:
    sol = solve_ivp(d_system_2, t_span, init_cond, args=params, dense_output=True)
    t = np.linspace(t_span[0], t_span[1], 50 * (t_span[1] - t_span[0]))
    y = sol.sol(t)
    plt.plot(y[0], y[1], label=f'Initial condition: {init_cond}')

# set the labels and title
plt.xlim(0, 3)
plt.ylim(0, 3)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Phase portrait of the dynamical system')
plt.legend()
plt.grid()
plt.savefig('Lotka-Volterra2.png', dpi=300)
plt.show()
```

```python
# define the system

def d_system_3(t, y, a, b, c, d, e, f):
    x, y_ = y
    dx_dt = x * (a - b * x - c * y_)
    dy_dt = y_ * (d - e * x - f * y_)
    return [dx_dt, dy_dt]
```

```python
# set parameters

a, b, c, d, e, f = 1, 2, 3, 4, 5, 6
params = (a, b, c, d, e, f)
```

```python
# define time span and initial conditions

t_span = (0, 50)
initial_conditions = [
    [0.125, 0.125],
    [0.25, 0.25],
    [0.5, 0.5],
    [1, 1],
]
```

```python
# plot the trajectories

plt.figure(figsize=(8, 6))
for init_cond in initial_conditions:
    sol = solve_ivp(d_system_3, t_span, init_cond, args=params, dense_output=True)
    t = np.linspace(t_span[0], t_span[1], 50 * (t_span[1] - t_span[0]))
    y = sol.sol(t)
    plt.plot(y[0], y[1], label=f'Initial condition: {init_cond}')

# plot the lines
x_range = np.linspace(0, 3, 300)
y1_range = (a - b * x_range) / c
y2_range = (d - e * x_range) / f
plt.plot(x_range, y1_range, '--', label='a - bx - cy = 0')
plt.plot(x_range, y2_range, '--', label='d - ex - fy = 0')

# set the labels and title

plt.xlim(0, 1)
plt.ylim(0, 1)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Phase portrait of the dynamical system')
plt.legend()
plt.grid()
plt.savefig('Lotka-Volterra3.png', dpi=300)
plt.show()
```

```python
# define the system

def d_system_4(t, y):
    x, y_ = y
    dx_dt = x - y_ - x * (x ** 2 + y_ ** 2)
    dy_dt = x + y_ - y_ * (x ** 2 + y_ ** 2)
    return [dx_dt, dy_dt]
```

```python
# import libs

import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
from mpl_toolkits.mplot3d import Axes3D
```

```python
# define the system

def d_system_3d(t, x, alpha, beta):
    x1, x2, x3 = x
    dx1_dt = x1 * (1 - x1 - alpha * x2 - beta * x3)
    dx2_dt = x2 * (1 - beta * x1 - x2 - alpha * x3)
    dx3_dt = x3 * (1 - alpha * x1 - beta * x2 - x3)
    return [dx1_dt, dx2_dt, dx3_dt]
```

```python
# set parameters

alpha, beta = 1.6, 0.8
params = (alpha, beta)
```

```python
# define time span and initial conditions

t_span = (0, 100)
initial_conditions = [
    [0.2, 0.4, 0.4],
    [0.4, 0.3, 0.3],
    [0.6, 0.3, 0.1],
    [0.1, 0.5, 0.4]
]
```

```python
# plot the trajectories

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

for init_cond in initial_conditions:
    sol = solve_ivp(d_system_3d, t_span, init_cond, args=params, dense_output=True, max_step=0.01)
    t = np.linspace(t_span[0], t_span[1], 300 * (t_span[1] - t_span[0]))
    y = sol.sol(t)
    ax.plot(y[0], y[1], y[2], label=f'Initial condition: {init_cond}', linewidth=0.5)

# set the labels and title

ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_zlim(0, 1)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('x3')
ax.set_title('3D Phase portrait of the dynamical system')

# set the view angle
elevation_angle = 30
azimuth_angle = 45
ax.view_init(elevation_angle, azimuth_angle)

ax.legend()
plt.savefig('3d-dynamical-system1.png', dpi=300)
plt.show()
```
